```python
import os
import pickle
from keyboard_phenotype import KeyboardPhenotype
from kle.kle_model import Serial, FingerName
from keyboard_genotypes import LAYOUT_DATA
from kle.kle_renderer import show_keyboard, render_keyboard

try:
    data_file = './processed/markov_chains.pkl'
    if not os.path.exists(data_file):
        print(f"ERROR: Data file not found: {data_file}")
        data = {}
    else:
        with open(data_file, 'rb') as f:
            data = pickle.load(f)
        print(f"✅ Successfully loaded data from {data_file}")
except Exception as e:
    print(f"ERROR loading data: {e}")
    data = {}
```

```python

with open('kle_keyboards/ansi_60_percent_hands.json', 'r') as f:
    keyboard = Serial.parse(f.read())

keyboard = KeyboardPhenotype(keyboard, {})

keyboard.select_remap_keys(LAYOUT_DATA['qwerty'])
keyboard.remap_to_keys(LAYOUT_DATA['asset'])
# keyboard.fitness(data['simple_wikipedia'])
show_keyboard(keyboard.get_phisical_keyboard())
```

```python
import matplotlib.pyplot as plt

# Collect fitness values
fitness_values = []
layout_keys = []

for key, value in LAYOUT_DATA.items():
    print(key)
    keyboard.remap_to_keys(value)
    fitness = keyboard.fitness(data['simple_wikipedia'])  # Store the returned fitness value
    fitness_values.append(fitness)
    layout_keys.append(key)
    show_keyboard(keyboard.get_phisical_keyboard())
```

```python

# Plot the fitness values in a bar chart
plt.figure(figsize=(10, 6))
plt.bar(layout_keys, fitness_values)
plt.xlabel('Layout')
plt.ylabel('Fitness Value')
plt.title('Fitness Values for Different Layouts')
plt.xticks(rotation=45)  # Rotate x-axis labels if there are many layouts
plt.tight_layout()  # Adjust layout to prevent label cutoff
plt.show()
```

```python
import matplotlib.pyplot as plt

# Blacklist specific dataset
BLACKLISTED_DATASET = ['catrigratis', 'gutenberg', 'the_algorithms_code']
DATASETS = data

# Collect fitness values for all layouts and datasets (except blacklisted)
results = {}

for layout_key, layout_value in LAYOUT_DATA.items():
    print(f"Processing layout: {layout_key}")
    results[layout_key] = {}
    
    # Process each dataset except the blacklisted one
    for dataset_name, dataset_data in DATASETS.items():
        print(f"processing dataset {dataset_name}")
        if dataset_name not in BLACKLISTED_DATASET:  # Skip blacklisted dataset
            keyboard.remap_to_keys(layout_value)
            fitness = keyboard.fitness(dataset_data)
            results[layout_key][dataset_name] = fitness
            print(f"  {dataset_name}: {fitness}")
        else:
            print(f"  Skipping blacklisted dataset: {dataset_name}")
    
    show_keyboard(keyboard.get_phisical_keyboard())

# Create bar charts for each non-blacklisted dataset
for dataset_name in DATASETS.keys():
    if dataset_name not in BLACKLISTED_DATASET:
        layout_keys = list(LAYOUT_DATA.keys())
        fitness_values = [results[layout_key][dataset_name] for layout_key in layout_keys]
        
        plt.figure(figsize=(10, 6))
        plt.bar(layout_keys, fitness_values)
        plt.xlabel('Layout')
        plt.ylabel('Fitness Value')
        plt.title(f'Fitness Values for Different Layouts - Dataset: {dataset_name}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

# Grouped bar chart for all non-blacklisted datasets
non_blacklisted_datasets = [name for name in DATASETS.keys() if name not in BLACKLISTED_DATASET]

fig, ax = plt.subplots(figsize=(12, 8))
x = range(len(LAYOUT_DATA))
width = 0.2

for i, dataset_name in enumerate(non_blacklisted_datasets):
    fitness_values = [results[layout_key][dataset_name] for layout_key in LAYOUT_DATA.keys()]
    ax.bar([pos + i * width for pos in x], fitness_values, width, label=dataset_name)

ax.set_xlabel('Layout')
ax.set_ylabel('Fitness Value')
ax.set_title('Fitness Values for Different Layouts Across Multiple Datasets (Excluding catrigratis)')
ax.set_xticks([pos + width * (len(non_blacklisted_datasets)-1) / 2 for pos in x])
ax.set_xticklabels(LAYOUT_DATA.keys())
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
